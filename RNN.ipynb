```
class encoder:
  def text_encoding(self, docs):
    t = Tokenizer()
    t.fit_on_texts(docs)
    encoded_docs = t.texts_to_sequences(docs)
    padded_docs = pad_sequences(encoded_docs, maxlen=15, padding='post')
    return padded_docs
  def label_encoding(self, labels):
    negative = [1, 0, 0]
    neutral = [0, 1, 0]
    positive = [0, 0, 1]
    n = len(labels)
    for i in range(n):
      if(labels[i] == 0):
        labels[i] = copy.deepcopy(negative)
      elif(labels[i] == 1):
        labels[i] = copy.deepcopy(neutral)
      elif(labels[i] == 2):
        labels[i] = copy.deepcopy(positive)
    return labels
```
